```python
# 선그래프 - 월 기준 
```

```python
df['월'].value_counts().sort_index().plot(
    figsize =(15,4), grid = True)
plt.axhline(3200, linestyle  =":" , color = "red")
```

```python
# 월별 시각화 - 최대 3634억  - 5월 
month_case = df['월'].value_counts().sort_index()
g = month_case.plot.bar(rot=0,  figsize =(15,4))

for i in range(len(month_case)):
    month_case_i = month_case.iloc[i]
    if month_case_i > 3200:
        g.text(x = month_case.index[i] - 1.2, y = month_case_i , s = month_case_i)
```

```python
# 가장 많이 팔린 시간대 
df['시간'].value_counts().sort_index().plot(
    figsize =(15,4), grid = True)
plt.axhline(3200, linestyle  =":" , color = "red")
```

```python
# 시간 시각화 - 최대 22시 3644억
hour_case = df['시간'].value_counts().sort_index()
g = hour_case.plot.bar(rot=0,  figsize =(15,4))

for i in range(len(hour_case)):
    hour_case_i = hour_case.iloc[i]
    if hour_case_i > 2000:
        g.text(x = hour_case.index[i] - 1.2, y = hour_case_i , s = hour_case_i)
```

```python
# 많이 팔린 주 
df['주'].value_counts().sort_index().plot(
    figsize =(15,4), grid = True)
plt.axhline(800, linestyle  =":" , color = "red")
```

```python
# 주 시각화 - 최대 많이 판 주 = 913억 
week_case = df['주'].value_counts().sort_index()
g = week_case.plot.bar(rot=0, figsize =(15,4))

for i in range(len(week_case)):
    week_case_i = week_case.iloc[i]
    if week_case_i > 800:
        g.text(x = week_case.index[i] - 1.2, y = week_case_i , s = week_case_i)
```

```python
# 월일의 데이터 생성
df['월일'] = df['방송일시'].astype(str).map(lambda x : x[-14:]) # 문자열로 바꾸고 뒤에서 부터 출력하여 년도 제외
day_count = df['월일'].value_counts().sort_index()
```
